```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Read CSV

```python
cats = pd.read_csv("https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2023/2023-01-31/cats_uk_reference.csv")
```

# Exploratory Data Analysis

## Number of Cats

```python
n_catscount = cats.value_counts("n_cats").reset_index(name = "count")
sns.barplot(n_catscount, x = "n_cats", y = "count")
plt.title("Number of Cats per Household")
plt.xlabel("Number of Cats")
plt.ylabel("Number of Households")
```

## Prey per Year

```python
prey_per_year = cats.value_counts("prey_p_month").reset_index(name = "count")
prey_per_year["prey_p_month"] = (prey_per_year["prey_p_month"]*12).astype(int)
sns.barplot(prey_per_year, x = "prey_p_month", y = "count")
plt.xlabel("Prey per Year")
plt.ylabel("Number of Cats")
plt.title("Number of Prey Caught per Year")
```

## Number of Hours Indoors

```python
indoors = cats.value_counts("hrs_indoors").reset_index(name = "count")
sns.barplot(indoors, x = "hrs_indoors", y = "count")
plt.title("Number of Hours Spent Indoors")
plt.xlabel("Hours")
plt.ylabel("Number of Hours")
```

## Cat Gender

```python
gender = cats.value_counts("animal_sex").reset_index(name = "count")
sns.barplot(gender, x = "animal_sex", y = "count")
plt.title("Cat Gender")
plt.ylabel("Number of Cats")
plt.xlabel("Gender")
```

## Cat Age

```python
ages = cats.value_counts("age_years").reset_index(name = "count") \
    .sort_values("count")
ages["age_years"] = (ages["age_years"]).astype(int)
sns.barplot(ages, x= "age_years", y = "count")
plt.title("Cat Age Distribution")
plt.xlabel("Age")
plt.ylabel("Number of Cats")
```
