```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Read CSV

```python
cats = pd.read_csv("https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2023/2023-01-31/cats_uk_reference.csv")
```

# Exploratory Data Analysis

```python
n_catscount = cats.value_counts("n_cats").reset_index(name = "count")
sns.barplot(n_catscount, x = "n_cats", y = "count")
plt.title("Number of Cats per Household")
plt.xlabel("Number of Cats")
```
